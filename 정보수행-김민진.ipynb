```python
# csv 모듈 가져오기
import csv

# CSV 파일 열어서 f 라는 변수로 할당하기
f = open('노인복지_생활시설_수_및_생활현황_19942004년__20180611084232.csv')

# 할당된 f 파일을 csv 읽어서 data 에 담기
data = csv.reader(f)

# 무료양로시설 숫자를 더할 변수 s를 만들고 0으로 초기화하기
s = 0

# 무료노인요양시설 숫자를 더할 변수 p 를 만들고 0으로 초기화
p = 0

# 실비 및 유료시설 숫자를 더할 변수 d 를 만들고 0으로 초기화
d = 0

# csv 파일의 내용을 줄 단위로 읽어서 row 라는 list 객체에 담고, 파일끝까지 읽기
for row in data :

    # 한줄 읽은 csv 파일의 두번째 행(컬럼)의 값이 '무료양로시설' 이면 아래 명령 실행하기
    if row[1] == '무료양로시설':
        # 무료양로시설 숫자 (csv 파일의 세번째 행(컬럼) 값) 를 s 에 더하기 --> 줄마다 나오는 무료양로시설을 모두 더하기
        s = s + int(row[2])
        # int(row[2]) : row[2]의 값은 문자열이므로 연산에 사용할 수 없다. 따라서 정수형으로 변환해야 한다.

    # 한줄 읽은 csv 파일 두번째 행(컬럼)의 값이 '무료노인요양시설' 이면 아래 명령 실행하기
    elif row[1] == '무료노인요양시설':
        # 무료노인요양시설 숫자 (csv 파일의 세번째 행(컬럼) 값) 를 s 에 더하기 --> 줄마다 나오는 무료노인요양시설을 모두 더하기
        p = p + int(row[2])

    # 한줄 읽은 csv 파일 두번째 행(컬럼)의 값이 '실비 및 유료시설' 이면 아래 명령 실행하기
    elif row[1] == '실비 및 유료시설':
        # 실비 및 유료시설 숫자 (csv 파일의 세번째 행(컬럼) 값) 를 s 에 더하기 --> 줄마다 나오는 실비 및 유료시설을 모두 더하기
        d = d + int(row[2])

print('전국 무료양로시설 합계:', s)
print('전국 무료노인요양시설 합계:', p)
print('전국 실비 및 유료시설 합계:', d)
```

```python
import csv

f = open('정보 자료 진짜 (1).csv')

data = csv.reader(f)
 
s=0

# csv 파일을 읽어 들이면 data 라는 리스트 객체에 담김, 전체 파일의 내용을 Sort(정렬)하는 데 사용할 리스트를 생성
# 이 프로그램에서는 정렬은 없이 저장만 함
llst = []

for row in data:
    lst = []
    if row[0] != '':
        lst.append(row[0])
        for n in range(1, len(row), 2):
            if int(row[n]) == 0 or int(row[n+1]) == 0:
                lst.append(0)
            else:
                s = int(row[n+1]) / int(row[n])
                lst.append(s)
        print(lst)
        llst.append(lst)
```

```python
import csv
f = open('정보 자료 진짜 (1).csv')
data = csv.reader(f)

s = 0
llst = []

for row in data:

    lst = []
    if row[0] != '':
        # s = int(row[2])/int(row[1])
        lst.append(row[0])
        for n in range(1, len(row), 2):
            if int(row[n]) == 0 or int(row[n + 1]) == 0:
                lst.append(0)
            else:
                s = int(row[n]) / int(row[n + 1])
                lst.append(s)
        # print (s)
        # print(lst)
        llst.append(lst)

# 파일의 내용을 계산하여 저장한 리스트 객체 llst 를 정렬(sort) 함, 정렬 기준은 llst (리스트 객체)의 두번째 행(컬럼)의 값을 기준으로 오름차순으로 정렬함
# sorted 라는 파이썬 내장 함수를 이용하였고, 정렬 기준은 람다식으로 표현하여 직접 기준 값을 정의함
print(sorted(llst, key=lambda x: llst[1]))
```
