```python
import csv

f = open('정보 자료 진짜(1)')

data = csv.reader(f)

print('!각 계산 결과는 내림을 했습니다!')

print('(지역 예시) 서울 부산 대구 인천 광주 대전 울산 세종 경기 강원 충북 충남 전북 전남 경북 경남 제주')

print('(정보 예시) 양로시설, 노인공동생활가정, 노인복지주택, 방문요양서비스')

n = input('원하는 지역을 입력해주세요 : ')

m = input('알고 싶은 정보를 입력해주세요 : ')

    

for row in data :

    if n == row[0] :

 

        if row[8] == '0':

            print(row[0] + ' 지역의 방문요양서비스 관련 데이터가 존재하지 않습니다.')

        

        if m in '양로시설' : 

            print(row[0] + ' 지역에서 한 양로시설 당 수용 인원 : ' +str((int(float(row[2])/float(row[1])))))

    

        if m in '노인공동생활가정' :

            if row[4] == '0' :

                 print(row[0] + ' 지역의 노인공동생활가정 관련 데어터가 존재하지 않습니다.')

            else :

                print(row[0] + ' 지역에서 한 노인공동생활가정의 인원 : ' +str(int(float(row[4])/float(row[3]))))

        

        if m in '노인복지주택' :

            if row[6] == '0':

                print(row[0] + ' 지역의 노인복지주택 관련 데이터가 존재하지 않습니다.')

            else :

                print(row[0] + ' 지역에서 한 노인복지주택의 입주 인원 : ' +str(int(float(row[6])/float(row[5]))))

        

        if m in '방문요양서비스' :

            if row[8] == '0':

                print(row[0] + ' 지역의 방문요양서비스 관련 데이터가 존재하지 않습니다.')

            else :

                print(row[0] + ' 지역에서 한 방문요양서비스 당 인원 : ' +str(int(float(row[8])/float(row[7]))))

            
```
